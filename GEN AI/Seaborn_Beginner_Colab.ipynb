```python
# Step 1: Import libraries
# Seaborn works on top of Matplotlib and Pandas
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

print("Seaborn version:", sns.__version__)
```

## ðŸ“Š Our One Example: Student Marks (Table)

We use a Pandas DataFrame (like Excel).


```python
# Create student marks DataFrame
data = {
    "Name": ["Arun", "Beena", "Charan", "Divya", "Esha"],
    "Math": [78, 66, 90, 55, 80],
    "Science": [85, 74, 88, 60, 72],
    "English": [92, 81, 95, 58, 77]
}

df = pd.DataFrame(data)
df
```

## ðŸ“Š Bar Plot (Average Marks per Subject)

Bar plots are great for **comparison**.


```python
# Convert data to long format for Seaborn
df_long = df.melt(id_vars="Name", var_name="Subject", value_name="Marks")

sns.barplot(data=df_long, x="Subject", y="Marks")
plt.title("Average Marks per Subject")
plt.show()
```

## ðŸ”¢ Count Plot (Number of Students)

Count plot simply **counts values**.


```python
sns.countplot(data=df_long, x="Subject")
plt.title("Number of Marks Entries per Subject")
plt.show()
```

## ðŸ“¦ Box Plot (Understand Score Spread)

Box plots show:
- Minimum
- Maximum
- Average range


```python
sns.boxplot(data=df_long, x="Subject", y="Marks")
plt.title("Marks Distribution per Subject")
plt.show()
```
