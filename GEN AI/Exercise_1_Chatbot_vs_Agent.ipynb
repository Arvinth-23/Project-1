# Exercise 1: Chatbot vs Agent

This notebook demonstrates the **conceptual difference between a Chatbot and an Agent**.

**Objective:** Understand why Agentic AI is different from prompt-based chatbots.

## Chatbot Example

This section presents a simple chatbot implementation. Chatbots typically respond directly to user input based on pre-defined rules or patterns, without maintaining a long-term plan or performing complex actions.

```python
print('=== CHATBOT MODE ===')

def chatbot(user_input):
    """Simulates a basic chatbot response."""
    return f"Chatbot response to '{user_input}': Here is some information."

print(chatbot('Plan my semester'))
```

## Agent Example

This section demonstrates an AI agent. Unlike a chatbot, an agent can break down a complex goal into a series of steps, execute those steps, and often interact with external tools or systems to achieve the goal.

```python
print('\n=== AGENT MODE ===')

def agent(user_goal):
    """Simulates an AI agent breaking down a goal into steps."""
    steps = [
        'Understand goal',
        'Check academic calendar',
        'Identify required courses',
        'Create semester plan',
        'Ask user for confirmation'
    ]
    print(f'Agent received goal: {user_goal}')
    for step in steps:
        print(f'Agent step: {step}')
    return 'Agent completed planning with actions.'

print(agent('Plan my semester'))
```
