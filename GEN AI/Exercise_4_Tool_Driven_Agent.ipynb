# Exercise 4: Tool-Driven Agent – Think vs Act

This notebook demonstrates **how an agent decides WHEN to think and WHEN to use tools**.

**Objective:** Understand why tools are mandatory for reliable Agentic AI systems.

## Scenario
Goal: **Answer a student query about exam schedule accurately**

The agent must decide whether to:
- Answer directly (think)
- Call a tool (act)

## 1️⃣ Tool Definition (Simulated)
This tool represents an official academic calendar system.

```python
def exam_schedule_tool():
    return "Mid-sem exams start on 10-Aug-2026"
```

## 2️⃣ Agent Decision Logic
The agent checks whether the question requires **verified data**.

```python
def tool_driven_agent(question):
    print("User Question:", question)
    
    if "exam" in question.lower():
        print("Decision: This requires official data → Use Tool")
        answer = exam_schedule_tool()
    else:
        print("Decision: General question → Reason internally")
        answer = "Here is a general explanation."
    
    return answer

print(tool_driven_agent("When are the mid-sem exams?"))
```

## 3️⃣ Before vs After Tools

| Without Tools | With Tools |
|--------------|------------|
| Guessing answers | Verifying facts |
| Hallucination risk | High accuracy |
| Not auditable | Auditable |
