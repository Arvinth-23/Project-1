```python
# Step 1: Import Matplotlib
# Matplotlib is used to draw charts and graphs
import matplotlib.pyplot as plt

print("Matplotlib imported successfully")
```

## ðŸ“š Our One Example: Student Marks

We will draw graphs using these marks.


```python
# Student names and their Math marks
students = ["Arun", "Beena", "Charan", "Divya", "Esha"]
math_marks = [78, 66, 90, 55, 80]

students, math_marks
```

## ðŸ“ˆ Line Chart

A line chart shows **change or trend**.


```python
plt.plot(students, math_marks)
plt.title("Math Marks - Line Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```

## ðŸ“Š Bar Chart

A bar chart is best for **comparing values**.


```python
plt.bar(students, math_marks)
plt.title("Math Marks - Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

## ðŸŸ¢ Adding Colors and Markers

We can make charts more attractive and readable.


```python
plt.plot(students, math_marks, marker='o', linestyle='--')
plt.title("Math Marks with Markers")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```
